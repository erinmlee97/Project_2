```python
#Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
starships= pd.read_csv("../data/starships.csv")
starships
```

```python
hyperdrive_df = starships[["name", "hyperdrive_rating"]]
```

```python
for index, row in hyperdrive_df.iterrows():
    if row['hyperdrive_rating'] == "unknown":
        hyperdrive_df.drop(index, inplace=True)
```

```python
for index, row in hyperdrive_df.iterrows():
    if row['name'] == "unknown":
        hyperdrive_df.drop(index, inplace=True)
```

```python
hyperdrive_df['hyperdrive_rating'] = pd.to_numeric(hyperdrive_df['hyperdrive_rating'])
```

```python
hyperdrive_df.dtypes
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(hyperdrive_df))
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(10,5))
plt.bar(x_axis, hyperdrive_df["hyperdrive_rating"], color='lightblue', alpha=0.5, align="center")
plt.xticks(tick_locations, hyperdrive_df["name"], rotation="vertical")
plt.xlabel("Starships")
plt.ylabel("Hyperdrive Rating")
plt.title("Hyperdrive Rating by Starship")
plt.tight_layout()
plt.show()
plt.savefig("../images/hyperdrive.png")
```
