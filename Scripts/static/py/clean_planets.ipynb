```python
import pandas as pd
```

```python
planets2 = pd.read_csv("../data/planets.csv")
planets2
```

```python
for index, row in planets2.iterrows():
    if row['orbital_period'] == "unknown":
        planets2.drop(index, inplace=True)
```

```python
for index, row in planets2.iterrows():
    if row['rotation_period'] == "unknown":
        planets2.drop(index, inplace=True)
```

```python
for index, row in planets2.iterrows():
    if row['population'] == "unknown":
        planets2.drop(index, inplace=True)
```

```python
for index, row in planets2.iterrows():
    if row['diameter'] == "unknown":
        planets2.drop(index, inplace=True)
```

```python
planets2
```

```python
for index, row in planets2.iterrows():
    if row['gravity'] == "unknown":
        planets2.drop(index, inplace=True)
```

```python
for index, row in planets2.iterrows():
    if row['surface_water'] == "unknown":
        planets2.drop(index, inplace=True)
```

```python
planets2.to_csv("../data/clean_planets.csv", index=False)
```
