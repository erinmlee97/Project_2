```python
#Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
planets= pd.read_csv("../data/planets.csv")
planets
```

```python
rotation_df = planets[["name", "rotation_period"]]
```

```python
for index, row in rotation_df.iterrows():
    if row['rotation_period'] == "unknown":
        rotation_df.drop(index, inplace=True)
```

```python
for index, row in rotation_df.iterrows():
    if row['name'] == "unknown":
        rotation_df.drop(index, inplace=True)
```

```python
rotation_df['rotation_period'] = pd.to_numeric(rotation_df['rotation_period'])
```

```python
rotation_df.dtypes
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(rotation_df))
tick_locations = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(10,5))
plt.bar(x_axis, rotation_df["rotation_period"], color='lightblue', alpha=0.5, align="center")
plt.xticks(tick_locations, rotation_df["name"], rotation="vertical")
plt.xlabel("Planets")
plt.ylabel("Rotation Period")
plt.title("Rotation Period per Planet")
plt.tight_layout()
plt.show()
plt.savefig("../images/rotation.png")
```
