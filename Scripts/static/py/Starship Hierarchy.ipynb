```python
import pandas as pd
```

```python
df = pd.read_csv("../data/starships.csv")
```

```python
starships_df = pd.DataFrame(df)
starships_df
```

```python
starships_df2 = starships_df.drop(df.index[8])
starships_df2
```

```python
starships_df3 = starships_df2.drop(df.index[13])
starships_df3
```

```python
starships_df4 = starships_df3.drop(df.index[24])
starships_df4
```

```python
starships_df5 = starships_df4.drop(df.index[17])
starships_df5
```

```python
starships_df6 = starships_df5.drop(df.index[27])
starships_df6
```

```python
for index, row in starships_df6.iterrows():
    if row['max_atmosphering_speed'] == "unknown":
        starships_df6.drop(index, inplace=True)
```

```python
for index, row in starships_df6.iterrows():
    if row['hyperdrive_rating'] == "unknown":
        starships_df6.drop(index, inplace=True)
```

```python
for index, row in starships_df6.iterrows():
    if row['MGLT'] == "unknown":
        starships_df6.drop(index, inplace=True)
```

```python
for index, row in starships_df6.iterrows():
    if row['passengers'] == "unknown":
        starships_df6.drop(index, inplace=True)
```

```python
starships_df6.to_csv("../data/starship_2.csv")
```
